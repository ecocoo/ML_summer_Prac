### 1.1 Создание тензоров

```python
# Создайте следующие тензоры:
# - Тензор размером 3x4, заполненный случайными числами от 0 до 1
# - Тензор размером 2x3x4, заполненный нулями
# - Тензор размером 5x5, заполненный единицами
# - Тензор размером 4x4 с числами от 0 до 15 (используйте reshape)
```

```python
tensor_1 = torch.rand((3, 4))
tensor_2 = torch.zeros((2,3,4))
tensor_3 = torch.ones((5,5))
tensor_4 = torch.arange(16).reshape(4, 4)
print(tensor_1)
print(tensor_2)
print(tensor_3)
print(tensor_4)
```

### 1.2 Операции с тензорами 

```python
# Дано: тензор A размером 3x4 и тензор B размером 4x3
# Выполните:
# - Транспонирование тензора A
# - Матричное умножение A и B
# - Поэлементное умножение A и транспонированного B
# - Вычислите сумму всех элементов тензора A
```

```python
A = torch.rand((3,4))
B = torch.rand((4,3))
print(f'Тензор A {A}')
print(f'Тензор B {B}')
transpose_A = A.T
print(f'Транспонирование тензора A {A.T}')
print(f'Матричное умножение {A@B}')
print(f'Поэлементное умножение {A * B.T}')
print(f'Сумма всех элементов тензора A {A.sum()}')
```

### 1.3 Индексация и срезы

```python
# Создайте тензор размером 5x5x5
# Извлеките:
# - Первую строку
# - Последний столбец
# - Подматрицу размером 2x2 из центра тензора
# - Все элементы с четными индексами
```

```python
tensor = torch.rand((5,5,5))
print(tensor)
print(f'Первая строка {tensor[0,0,:]}')
print(f'Последний столбец {tensor[-1,:,-1]}')
print(f'Подматрицу размером 2x2 из центра тензора {tensor[2, 2:4, 2:4]}')
print(f'Все элементы с четными индексами {tensor[::2,::2,::2]}')
```

### 1.4 Работа с формами

```python
# Создайте тензор размером 24 элемента
# Преобразуйте его в формы:
# - 2x12
# - 3x8
# - 4x6
# - 2x3x4
# - 2x2x2x3
```

```python
tensor_for_reshape = torch.arange(24)
print(tensor_for_reshape)
print(tensor_for_reshape.reshape((2,12)))
print(tensor_for_reshape.reshape((3,8)))
print(tensor_for_reshape.reshape((4,6)))
print(tensor_for_reshape.reshape((2,3,4)))
print(tensor_for_reshape.reshape((2,2,2,3)))
```
