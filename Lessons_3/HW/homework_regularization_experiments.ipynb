```python
import torch
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import get_cifar_loaders
from models import FullyConnectedModel
from utils import plot_training_history, count_parameters
import numpy as np
import torch.optim as optim
from tqdm import tqdm
import torch.nn as nn

def run_epoch(model, data_loader, criterion, optimizer=None, device='cpu', is_test=False):
    if is_test:
        model.eval()
    else:
        model.train()
    
    total_loss = 0
    correct = 0
    total = 0
    
    for batch_idx, (data, target) in enumerate(tqdm(data_loader)):
        data, target = data.to(device), target.to(device)
        
        if not is_test and optimizer is not None:
            optimizer.zero_grad()
        
        output = model(data)
        loss = criterion(output, target)
        
        if not is_test and optimizer is not None:
            loss.backward()
            optimizer.step()
        
        total_loss += loss.item()
        pred = output.argmax(dim=1, keepdim=True)
        correct += pred.eq(target.view_as(pred)).sum().item()
        total += target.size(0)
    
    return total_loss / len(data_loader), correct / total

# Переопределим функцию train_model
def train_model(model, train_loader, test_loader, epochs=10, lr=0.001, weight_decay=0, device='cpu'):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    
    train_losses, train_accs = [], []
    test_losses, test_accs = [], []
    
    for epoch in range(epochs):
        train_loss, train_acc = run_epoch(model, train_loader, criterion, optimizer, device, is_test=False)
        test_loss, test_acc = run_epoch(model, test_loader, criterion, None, device, is_test=True)
        
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_losses.append(test_loss)
        test_accs.append(test_acc)
        
        print(f'Epoch {epoch+1}/{epochs}:')
        print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}')
        print(f'Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}')
        print('-' * 50)
    
    return {
        'train_losses': train_losses,
        'train_accs': train_accs,
        'test_losses': test_losses,
        'test_accs': test_accs
    }

# Получаем данные CIFAR
train_loader, test_loader = get_cifar_loaders(batch_size=64)

# общая архитектура
base_config = {
    "input_size": 3072,
    "num_classes": 10,
    "layers": [
        {"type": "linear", "size": 256}, {"type": "relu"},
        {"type": "linear", "size": 128}, {"type": "relu"},
        {"type": "linear", "size": 64}, {"type": "relu"},
        {"type": "linear", "size": 10}
    ]
}

# Список конфигураций
configs = [
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"], "regularization": "none", "weight_decay": 0},  # Без регуляризации
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"] + [{"type": "dropout", "rate": 0.1}], "regularization": "dropout", "weight_decay": 0},  # Dropout 0.1
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"] + [{"type": "dropout", "rate": 0.3}], "regularization": "dropout", "weight_decay": 0},  # Dropout 0.3
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"] + [{"type": "dropout", "rate": 0.5}], "regularization": "dropout", "weight_decay": 0},  # Dropout 0.5
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"] + [{"type": "batch_norm"}], "regularization": "batch_norm", "weight_decay": 0},  # BatchNorm
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"] + [{"type": "dropout", "rate": 0.2}, {"type": "batch_norm"}], "regularization": "dropout_batchnorm", "weight_decay": 0},  # Dropout + BatchNorm
    {"input_size": 3072, "num_classes": 10, "layers": base_config["layers"], "regularization": "l2", "weight_decay": 0.01},  # L2 регуляризация
]

# обучение моделей и визуализация
def train_evaluate(configs):
    results = []
    for config in configs:
        print(f"Обучение модели с {config['regularization']}")

        model = FullyConnectedModel(
            input_size=config["input_size"],
            num_classes=config["num_classes"],
            layers=config["layers"]
        )

        start_time = time.time()  # Засекаем время начала

        # Обучение модели
        history = train_model(
            model, 
            train_loader, 
            test_loader, 
            epochs=10,
            lr=0.001,
            weight_decay=config["weight_decay"],
        )

        end_time = time.time()  # Засекаем время окончания
        training_time = end_time - start_time

        #точность на тестовом наборе
        test_accuracy = history['test_accs'][-1]

        results.append({
            "config": config,
            "test_accuracy": test_accuracy,
            "training_time": training_time,
            "history": history,
            "model": model,
            "regularization": config["regularization"]
        })

    return results

results = train_evaluate(configs)

# Визуализация результатов
def plot_results(results):
    plt.figure(figsize=(12, 8))
    
    for result in results:
        history = result["history"]
        label = result["regularization"]
        plt.plot(history['train_accs'], label=f"{label} - Обучающая")
        plt.plot(history['test_accs'], '--', label=f"{label} - Тестовая")
    
    plt.xlabel('Эпоха')
    plt.ylabel('Точность')
    plt.legend()
    plt.title('Точность по эпохам для разных методов регуляризации (CIFAR-10)')
    plt.grid(True)
    plt.show()

plot_results(results)

# Визуализация распределения весов
def plot_weight_distribution(model, regularization_method):
    for name, param in model.named_parameters():
        if 'weight' in name and 'layers.0' in name:
            weights = param.data.cpu().numpy().flatten()
            break
    
    plt.figure(figsize=(10, 6))
    sns.histplot(weights, kde=True)
    plt.title(f"Распределение весов - {regularization_method}")
    plt.xlabel("Значение веса")
    plt.ylabel("Частота")
    plt.grid(True)
    plt.show()

# Визуализация распределений весов для каждой модели
for result in results:
    plot_weight_distribution(result["model"], result["regularization"])

# Сводная таблица результатов
print("\nСводные результаты:")
print("Метод регуляризации".ljust(25), "Точность на тесте".ljust(20), "Время обучения")
print("-" * 60)
for result in results:
    reg = result["regularization"]
    acc = result["test_accuracy"]
    time_val = result["training_time"]
    print(f"{reg.ljust(25)} {str(round(acc, 4)).ljust(20)} {round(time_val, 2)} секунд")
```

```python
import torch
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import get_mnist_loaders
from models import FullyConnectedModel
from trainer import train_model
from utils import plot_training_history, count_parameters
import numpy as np
from torch import nn

# Переопределим класс модели для поддержки параметров
class AdvancedFullyConnectedModel(nn.Module):
    def __init__(self, layers_config):
        super().__init__()
        self.layers = nn.ModuleList()
        self.adaptive_params = {}
        
        for layer_spec in layers_config:
            layer_type = layer_spec['type']
            
            if layer_type == 'linear':
                self.layers.append(nn.Linear(layer_spec['in_features'], layer_spec['out_features']))
                
            elif layer_type == 'relu':
                self.layers.append(nn.ReLU())
                
            elif layer_type == 'dropout':
                self.layers.append(nn.Dropout(p=layer_spec['p']))
                if 'adaptive' in layer_spec and layer_spec['adaptive']:
                    self.adaptive_params[id(self.layers[-1])] = {
                        'initial_p': layer_spec['p'],
                        'current_epoch': 0
                    }
                    
            elif layer_type == 'batch_norm':
                momentum = layer_spec.get('momentum', 0.1)
                self.layers.append(nn.BatchNorm1d(layer_spec['features'], momentum=momentum))
                
    def forward(self, x):
        x = x.view(x.size(0), -1)
        for layer in self.layers:
            x = layer(x)
        return x
    
    def update_adaptive_params(self, epoch):
        """Обновление адаптивных параметров"""
        for layer in self.layers:
            if id(layer) in self.adaptive_params:
                params = self.adaptive_params[id(layer)]
                # уменьшаем dropout с каждой эпохой
                decay_rate = 0.95
                new_p = params['initial_p'] * (decay_rate ** (epoch // 2))
                layer.p = max(new_p, 0.1)
                params['current_epoch'] = epoch

# Получаем данные MNIST
train_loader, test_loader = get_mnist_loaders(batch_size=64)

# общая архитектура
base_config = [
    {"type": "linear", "in_features": 784, "out_features": 256},
    {"type": "relu"},
    {"type": "linear", "in_features": 256, "out_features": 128},
    {"type": "relu"},
    {"type": "linear", "in_features": 128, "out_features": 64},
    {"type": "relu"}
]

# Конфигурация
configs = [
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5, "adaptive": True},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Адаптивный Dropout"
    },
    {
        "layers": base_config + [
            {"type": "batch_norm", "features": 64, "momentum": 0.9},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "BatchNorm (momentum=0.9)"
    },
    {
        "layers": base_config + [
            {"type": "batch_norm", "features": 64, "momentum": 0.99},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "BatchNorm (momentum=0.99)"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.3},
            {"type": "batch_norm", "features": 64, "momentum": 0.9},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Dropout + BatchNorm"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5, "adaptive": True},
            {"type": "batch_norm", "features": 64, "momentum": 0.95},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Адаптивный Dropout + BatchNorm"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Фиксированный Dropout"
    }
]

# Модифицируем функцию train_model для поддержки адаптивности
def adaptive_train_model(model, train_loader, test_loader, epochs=10, lr=0.001, device='cpu'):
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    train_losses, train_accs = [], []
    test_losses, test_accs = [], []
    
    for epoch in range(epochs):
        if hasattr(model, 'update_adaptive_params'):
            model.update_adaptive_params(epoch)
        
        # процесс обучения
        model.train()
        total_loss = 0
        correct = 0
        total = 0
        
        for data, target in train_loader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()
            total += target.size(0)
        
        train_loss = total_loss / len(train_loader)
        train_acc = correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        
        # Тестирование
        model.eval()
        total_loss = 0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                loss = criterion(output, target)
                total_loss += loss.item()
                pred = output.argmax(dim=1, keepdim=True)
                correct += pred.eq(target.view_as(pred)).sum().item()
                total += target.size(0)
        
        test_loss = total_loss / len(test_loader)
        test_acc = correct / total
        test_losses.append(test_loss)
        test_accs.append(test_acc)
        
        print(f'Epoch {epoch+1}/{epochs}: Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}')
    
    return {
        'train_losses': train_losses,
        'train_accs': train_accs,
        'test_losses': test_losses,
        'test_accs': test_accs
    }

# обучение моделей и визуализация результатов
def train_evaluate(configs):
    results = []
    for config in configs:
        print(f"Обучение модели: {config['name']}")
        
        model = AdvancedFullyConnectedModel(config["layers"])
        
        start_time = time.time()
        history = adaptive_train_model(model, train_loader, test_loader, epochs=10)
        end_time = time.time()
        training_time = end_time - start_time
        
        test_accuracy = history['test_accs'][-1]
        
        results.append({
            "config": config,
            "test_accuracy": test_accuracy,
            "training_time": training_time,
            "history": history,
            "model": model,
            "name": config['name']
        })
    
    return results

results = train_evaluate(configs)

# Визуализация результатов
def plot_results(results):
    plt.figure(figsize=(12, 8))
    for result in results:
        history = result["history"]
        plt.plot(history['test_accs'], label=result["name"])
    
    plt.xlabel('Эпоха')
    plt.ylabel('Точность')
    plt.legend()
    plt.title('Сравнение адаптивных техник регуляризации')
    plt.grid(True)
    plt.show()

plot_results(results)

# Анализ влияния на разные слои
def analyze_layer_impact(model, model_name):
    print(f"\nАнализ модели: {model_name}")
    
    model.eval()
    layer_outputs = {}
    
    hooks = []
    for i, layer in enumerate(model.layers):
        def hook(module, input, output, idx=i):
            layer_outputs[idx] = output.detach()
        hooks.append(layer.register_forward_hook(hook))
    
    test_data, _ = next(iter(test_loader))
    with torch.no_grad():
        model(test_data[:1])
    
    for hook in hooks:
        hook.remove()
    
    # Визуализация распределений
    plt.figure(figsize=(15, 10))
    for i, outputs in layer_outputs.items():
        if outputs.numel() > 0:
            plt.subplot(3, 3, i+1)
            sns.histplot(outputs.cpu().numpy().flatten(), kde=True)
            plt.title(f"Слой {i}: {model.layers[i].__class__.__name__}")
            plt.xlabel("Значение активации")
    
    plt.tight_layout()
    plt.suptitle(f"Распределение активаций по слоям: {model_name}", y=1.02)
    plt.show()

for result in results:
    analyze_layer_impact(result["model"], result["name"])
```

```python
import torch
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import get_cifar_loaders
from models import FullyConnectedModel
from trainer import train_model
from utils import plot_training_history, count_parameters
import numpy as np
from torch import nn

# Переопределим класс модели для поддержки параметров
class AdvancedFullyConnectedModel(nn.Module):
    def __init__(self, layers_config):
        super().__init__()
        self.layers = nn.ModuleList()
        self.adaptive_params = {}
        
        for layer_spec in layers_config:
            layer_type = layer_spec['type']
            
            if layer_type == 'linear':
                self.layers.append(nn.Linear(layer_spec['in_features'], layer_spec['out_features']))
                
            elif layer_type == 'relu':
                self.layers.append(nn.ReLU())
                
            elif layer_type == 'dropout':
                self.layers.append(nn.Dropout(p=layer_spec['p']))
                if 'adaptive' in layer_spec and layer_spec['adaptive']:
                    self.adaptive_params[id(self.layers[-1])] = {
                        'initial_p': layer_spec['p'],
                        'current_epoch': 0
                    }
                    
            elif layer_type == 'batch_norm':
                momentum = layer_spec.get('momentum', 0.1)
                self.layers.append(nn.BatchNorm1d(layer_spec['features'], momentum=momentum))
                
    def forward(self, x):
        x = x.view(x.size(0), -1)
        for layer in self.layers:
            x = layer(x)
        return x
    
    def update_adaptive_params(self, epoch):
        """Обновление адаптивных параметров"""
        for layer in self.layers:
            if id(layer) in self.adaptive_params:
                params = self.adaptive_params[id(layer)]
                # уменьшаем dropout с каждой эпохой
                decay_rate = 0.95
                new_p = params['initial_p'] * (decay_rate ** (epoch // 2))
                layer.p = max(new_p, 0.1)
                params['current_epoch'] = epoch

# Получаем данные CIFAR
train_loader, test_loader = get_cifar_loaders(batch_size=64)

# общая архитектура
base_config = [
    {"type": "linear", "in_features": 3072, "out_features": 256},
    {"type": "relu"},
    {"type": "linear", "in_features": 256, "out_features": 128},
    {"type": "relu"},
    {"type": "linear", "in_features": 128, "out_features": 64},
    {"type": "relu"}
]

# Конфигурация
configs = [
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5, "adaptive": True},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Адаптивный Dropout"
    },
    {
        "layers": base_config + [
            {"type": "batch_norm", "features": 64, "momentum": 0.9},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "BatchNorm (momentum=0.9)"
    },
    {
        "layers": base_config + [
            {"type": "batch_norm", "features": 64, "momentum": 0.99},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "BatchNorm (momentum=0.99)"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.3},
            {"type": "batch_norm", "features": 64, "momentum": 0.9},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Dropout + BatchNorm"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5, "adaptive": True},
            {"type": "batch_norm", "features": 64, "momentum": 0.95},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Адаптивный Dropout + BatchNorm"
    },
    {
        "layers": base_config + [
            {"type": "dropout", "p": 0.5},
            {"type": "linear", "in_features": 64, "out_features": 10}
        ],
        "name": "Фиксированный Dropout"
    }
]

# Модифицируем функцию train_model для поддержки адаптивности
def adaptive_train_model(model, train_loader, test_loader, epochs=10, lr=0.001, device='cpu'):
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    train_losses, train_accs = [], []
    test_losses, test_accs = [], []
    
    for epoch in range(epochs):
        if hasattr(model, 'update_adaptive_params'):
            model.update_adaptive_params(epoch)
        
        # процесс обучения
        model.train()
        total_loss = 0
        correct = 0
        total = 0
        
        for data, target in train_loader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()
            total += target.size(0)
        
        train_loss = total_loss / len(train_loader)
        train_acc = correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        
        # Тестирование
        model.eval()
        total_loss = 0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                loss = criterion(output, target)
                total_loss += loss.item()
                pred = output.argmax(dim=1, keepdim=True)
                correct += pred.eq(target.view_as(pred)).sum().item()
                total += target.size(0)
        
        test_loss = total_loss / len(test_loader)
        test_acc = correct / total
        test_losses.append(test_loss)
        test_accs.append(test_acc)
        
        print(f'Epoch {epoch+1}/{epochs}: Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}')
    
    return {
        'train_losses': train_losses,
        'train_accs': train_accs,
        'test_losses': test_losses,
        'test_accs': test_accs
    }

# обучение моделей и визуализация результатов
def train_evaluate(configs):
    results = []
    
    for config in configs:
        print(f"\nОбучение модели: {config['name']}")
        
        model = AdvancedFullyConnectedModel(config["layers"])
        
        start_time = time.time()
        history = adaptive_train_model(
            model, 
            train_loader, 
            test_loader, 
            epochs=10,
        )
        end_time = time.time()
        training_time = end_time - start_time
        
        test_accuracy = history['test_accs'][-1]
        
        results.append({
            "config": config,
            "test_accuracy": test_accuracy,
            "training_time": training_time,
            "history": history,
            "model": model,
            "name": config['name']
        })
    
    return results

results = train_evaluate(configs)

# Визуализация результатов
def plot_results(results):
    plt.figure(figsize=(12, 8))
    for result in results:
        history = result["history"]
        plt.plot(history['test_accs'], label=result["name"])
    
    plt.xlabel('Эпоха')
    plt.ylabel('Точность')
    plt.legend()
    plt.title('Сравнение адаптивных техник регуляризации (CIFAR-10)')
    plt.grid(True)
    plt.show()

plot_results(results)

# Анализ влияния на разные слои
def analyze_layer_impact(model, model_name):
    print(f"Анализ модели: {model_name}")
    
    model.eval()
    layer_outputs = {}
    
    hooks = []
    for i, layer in enumerate(model.layers):
        def hook(module, input, output, idx=i):
            layer_outputs[idx] = output.detach().cpu()
        hooks.append(layer.register_forward_hook(hook))
    
    test_data, _ = next(iter(test_loader))
    with torch.no_grad():
        model(test_data[:1].to(model.layers[0].weight.device))
    
    for hook in hooks:
        hook.remove()
    
    # Визуализация распределений
    plt.figure(figsize=(15, 10))
    for i, outputs in layer_outputs.items():
        if outputs.numel() > 0:
            plt.subplot(3, 3, i+1)
            sns.histplot(outputs.numpy().flatten(), kde=True)
            plt.title(f"Слой {i}: {model.layers[i].__class__.__name__}")
            plt.xlabel("Значение активации")
    
    plt.tight_layout()
    plt.suptitle(f"Распределение активаций по слоям: {model_name}", y=1.02)
    plt.show()

for result in results:
    analyze_layer_impact(result["model"], result["name"])
```
