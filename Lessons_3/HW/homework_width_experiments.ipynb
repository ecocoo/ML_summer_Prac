```python
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import get_mnist_loaders
from models import FullyConnectedModel
from trainer import train_model
from utils import plot_training_history, count_parameters

# Получаем данные MNIST
train_loader, test_loader = get_mnist_loaders(batch_size=64)

# Параметры
width_options = [
    [64, 32, 16],   # Сужение
    [64, 64, 64],   # Постоянная ширина
    [16, 32, 64],   # Расширение
    [128, 256, 512] # Широкие слои
]

#количество слоев
depth_options = [1, 2, 3]

regularization_options = [None, "dropout", "batch_norm", "dropout_batch_norm"]

# Список всех конфигураций
configs = []

for depth in depth_options:
    for width in width_options:
        for reg in regularization_options:
            layers = []
            for i in range(depth):
                layers.append({"type": "linear", "size": width[i]})
                layers.append({"type": "relu"})
                
                if reg == "dropout":
                    layers.append({"type": "dropout", "rate": 0.2})
                elif reg == "batch_norm":
                    layers.append({"type": "batch_norm"})
                elif reg == "dropout_batch_norm":
                    layers.append({"type": "dropout", "rate": 0.2})
                    layers.append({"type": "batch_norm"})
            
            layers.append({"type": "linear", "size": 10})
            configs.append({"layers": layers, "depth": depth, "width": width, "reg": reg})

# Обучение и оценка
def train_evaluate(configs):
    results = []
    for config in configs:
        model = FullyConnectedModel(layers=config["layers"])
        start_time = time.time()

        history = train_model(model, train_loader, test_loader, epochs=5)

        end_time = time.time()
        training_time = end_time - start_time

        test_accuracy = history['test_accs'][-1]

        results.append({
            "config": config,
            "test_accuracy": test_accuracy,
            "training_time": training_time
        })
    
    return results

# grid search
results = train_evaluate(configs)

heatmap_data = np.zeros((len(depth_options), len(width_options)))

for i, depth in enumerate(depth_options):
    for j, width in enumerate(width_options):
        accuracies = [result["test_accuracy"] for result in results 
                     if result["config"]["depth"] == depth 
                     and result["config"]["width"] == width]
        if accuracies:
            heatmap_data[i, j] = np.mean(accuracies)

# Визуализация heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data, 
    xticklabels=[str(width) for width in width_options], 
    yticklabels=[str(depth) for depth in depth_options], 
    cmap="YlGnBu", 
    annot=True,
    fmt=".4f" 
)
plt.title("Heatmap")
plt.xlabel("Ширина")
plt.ylabel("Глубина")
plt.show()

# лучшая конфигурация
best_result = max(results, key=lambda x: x["test_accuracy"])
print("\nЛучшая конфигурация:")
print(f"Глубина: {best_result['config']['depth']} слоя")
print(f"Ширина: {best_result['config']['width']}")
print(f"Регуляризация: {best_result['config']['reg'] or 'Нет'}")
print(f"Точность на тесте: {best_result['test_accuracy']:.4f}")
print(f"Время обучения: {best_result['training_time']:.2f} секунд")
```

```python
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import get_cifar_loaders
from models import FullyConnectedModel
from trainer import train_model
from utils import plot_training_history, count_parameters
import torch

# Получаем данные CIFAR
train_loader, test_loader = get_cifar_loaders(batch_size=64)

# Базовые параметры
input_size = 3072
num_classes = 10

# Параметры
width_options = [
    [64, 32, 16],   # Сужение
    [64, 64, 64],   # Постоянная ширина
    [16, 32, 64],   # Расширение
    [128, 256, 512] # Широкие слои
]

# Количество слоев
depth_options = [1, 2, 3]

regularization_options = [None, "dropout", "batch_norm", "dropout_batch_norm"]

# Список всех конфигураций
configs = []

for depth in depth_options:
    for width in width_options:
        for reg in regularization_options:
            layers = []
            for i in range(depth):
                layers.append({"type": "linear", "size": width[i]})
                layers.append({"type": "relu"})
                
                if reg == "dropout":
                    layers.append({"type": "dropout", "rate": 0.2})
                elif reg == "batch_norm":
                    layers.append({"type": "batch_norm"})
                elif reg == "dropout_batch_norm":
                    layers.append({"type": "dropout", "rate": 0.2})
                    layers.append({"type": "batch_norm"})
            
            layers.append({"type": "linear", "size": num_classes})
            configs.append({
                "input_size": input_size,
                "num_classes": num_classes,
                "layers": layers, 
                "depth": depth, 
                "width": width, 
                "reg": reg
            })

# Обучение и оценка
def train_evaluate(configs):
    results = []
    
    for config in configs:
        model = FullyConnectedModel(
            input_size=config["input_size"],
            num_classes=config["num_classes"],
            layers=config["layers"]
        )
        
        start_time = time.time()
        
        history = train_model(
            model, 
            train_loader, 
            test_loader, 
            epochs=10,
        )
        
        end_time = time.time()
        training_time = end_time - start_time
        
        test_accuracy = history['test_accs'][-1]
        
        results.append({
            "config": config,
            "test_accuracy": test_accuracy,
            "training_time": training_time
        })
    
    return results

# Grid search
results = train_evaluate(configs)

heatmap_data = np.zeros((len(depth_options), len(width_options)))

for i, depth in enumerate(depth_options):
    for j, width in enumerate(width_options):
        accuracies = [result["test_accuracy"] for result in results 
                     if result["config"]["depth"] == depth 
                     and result["config"]["width"] == width]
        if accuracies:
            heatmap_data[i, j] = np.mean(accuracies)

# Визуализация heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data, 
    xticklabels=[str(width) for width in width_options], 
    yticklabels=[str(depth) for depth in depth_options], 
    cmap="YlGnBu", 
    annot=True,
    fmt=".4f" 
)
plt.title("Heatmap")
plt.xlabel("Ширина")
plt.ylabel("Глубина")
plt.show()

# Анализ лучшей конфигурации
best_result = max(results, key=lambda x: x["test_accuracy"])
print("\nЛучшая конфигурация:")
print(f"Глубина: {best_result['config']['depth']} слоя")
print(f"Ширина: {best_result['config']['width']}")
print(f"Регуляризация: {best_result['config']['reg'] or 'Нет'}")
print(f"Точность на тесте: {best_result['test_accuracy']:.4f}")
print(f"Время обучения: {best_result['training_time']:.2f} секунд")
```
