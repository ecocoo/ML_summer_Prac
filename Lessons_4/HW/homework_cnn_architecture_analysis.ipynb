```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import time
import matplotlib.pyplot as plt
from datasets import get_cifar_loaders
from trainer import train_model
from utils import count_parameters

# Конфигурация
BATCH_SIZE = 128
EPOCHS = 10
LR = 0.001
DEVICE = 'cpu'

train_loader, test_loader = get_cifar_loaders(batch_size=BATCH_SIZE)

# Модели с разными ядрами
class ConvKernelNet(nn.Module):
    def __init__(self, kernel_size, name="", use_1x1=False):
        super().__init__()
        padding = kernel_size // 2
        if use_1x1:
            self.conv1 = nn.Sequential(
                nn.Conv2d(3, 32, kernel_size=1),
                nn.ReLU(),
                nn.Conv2d(32, 64, kernel_size=kernel_size, padding=padding),
                nn.ReLU()
            )
        else:
            self.conv1 = nn.Sequential(
                nn.Conv2d(3, 64, kernel_size=kernel_size, padding=padding),
                nn.ReLU()
            )

        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(128 * 8 * 8, 256)
        self.fc2 = nn.Linear(256, 10)
        self.name = name

    def forward(self, x):
        x = self.pool(self.conv1(x))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# Поддержка равного числа параметров
def scale_channels(base_kernel_model):
    params = count_parameters(base_kernel_model)
    target = 315000
    ratio = (target / params) ** 0.5
    return ratio

kernels = {
    "k3": ConvKernelNet(3, "3x3"),
    "k5": ConvKernelNet(5, "5x5"),
    "k7": ConvKernelNet(7, "7x7"),
    "combo": ConvKernelNet(3, "1x1+3x3", use_1x1=True)
}

results = {}

# Обучение и анализ
for key, model in kernels.items():
    model = model.to(DEVICE)
    print(f"\Обучение модели {model.name} ({count_parameters(model)} параметров)")
    start = time.time()
    history = train_model(model, train_loader, test_loader, epochs=EPOCHS, lr=LR, device=DEVICE)
    end = time.time()
    history['train_time'] = end - start
    results[key] = {
        'model': model,
        'history': history,
        'params': count_parameters(model)
    }

# Визуализация
def visualize_first_layer(model, title):
    with torch.no_grad():
        x, _ = next(iter(train_loader))
        x = x[:1].to(DEVICE)
        acts = model.conv1(x).cpu().squeeze()
        num_filters = acts.shape[0]
        plt.figure(figsize=(12, 4))
        for i in range(min(num_filters, 10)):
            plt.subplot(2, 5, i + 1)
            plt.imshow(acts[i], cmap='viridis')
            plt.axis('off')
        plt.suptitle(f"Activations - {title}")
        plt.tight_layout()
        plt.show()

for key in kernels:
    visualize_first_layer(results[key]['model'], results[key]['model'].name)

# Анализ рецептивных полей
def compute_receptive_field(kernel_sizes):
    rf = 1
    for k in kernel_sizes:
        rf = rf + (k - 1)
    return rf

for key, model in kernels.items():
    if model.name == "1x1+3x3":
        rf = compute_receptive_field([1, 3, 3])
    else:
        rf = compute_receptive_field([model.conv1[0].kernel_size[0], 3])
    print(f" Рецептивное поле модели{model.name}: {rf}")

# Результаты
print("\nРезультаты:")
for key, res in results.items():
    acc = res['history']['test_accs'][-1]
    print(f"{res['model'].name}: параметры={res['params']}, итоговая точность на тесте={acc:.4f}, время оубчения={res['history']['train_time']:.2f}s")
```

```python
import time
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from datasets import get_cifar_loaders
from models import CNNWithResidual
from trainer import run_epoch
from utils import count_parameters

# Конфигурация
batch_size = 128
epochs = 10
lr = 0.001
device = 'cpu'

# Загрузка CIFAR-10
train_loader, test_loader = get_cifar_loaders(batch_size=batch_size)

# Разные по глубине CNN
class ShallowCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, 1, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1, 1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc = nn.Linear(64 * 8 * 8, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        return self.fc(x)

class MediumCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 32, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 128, 3, 1, 1), nn.ReLU()
        )
        self.pool = nn.AdaptiveAvgPool2d((4, 4))
        self.fc = nn.Linear(128 * 4 * 4, 10)

    def forward(self, x):
        x = self.conv(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

class DeepCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 32, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 128, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(128, 128, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(128, 256, 3, 1, 1), nn.ReLU()
        )
        self.pool = nn.AdaptiveAvgPool2d((4, 4))
        self.fc = nn.Linear(256 * 4 * 4, 10)

    def forward(self, x):
        x = self.conv(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)
        
# Инициализация моделей
models = {
    'ShallowCNN': ShallowCNN().to(device),
    'MediumCNN': MediumCNN().to(device),
    'DeepCNN': DeepCNN().to(device),
    'ResNet': CNNWithResidual(input_channels=3, num_classes=10).to(device)
}

results = {}

# Визуализация
def plot_feature_maps(model, name):
    model.eval()
    with torch.no_grad():
        x, _ = next(iter(train_loader))
        x = x.to(device)
        if hasattr(model, 'conv'):
            acts = model.conv(x[:1]).cpu()
        else:
            acts = model.conv1(x[:1]).cpu()
        plt.figure(figsize=(12, 4))
        for i in range(min(acts.shape[1], 8)):
            plt.subplot(2, 4, i+1)
            plt.imshow(acts[0, i], cmap='viridis')
            plt.axis('off')
        plt.suptitle(f"Карты признаков - {name}")
        plt.tight_layout()
        plt.show()

def plot_grad_flow(named_parameters):
    ave_grads = []
    layers = []
    for n, p in named_parameters:
        if p.requires_grad and p.grad is not None and "bias" not in n:
            layers.append(n)
            ave_grads.append(p.grad.abs().mean().item())
    plt.figure(figsize=(10, 4))
    plt.plot(ave_grads)
    plt.xticks(range(len(layers)), layers, rotation='vertical')
    plt.title("Поток градиентов")
    plt.tight_layout()
    plt.show()

# Обучение и анализ
for name, model in models.items():
    print(f"\n{name} ({count_parameters(model)} параметров)")
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()

    history = {'train_acc': [], 'test_acc': [], 'train_loss': [], 'test_loss': []}
    start = time.time()
    for epoch in range(epochs):
        train_loss, train_acc = run_epoch(model, train_loader, criterion, optimizer, device)
        test_loss, test_acc = run_epoch(model, test_loader, criterion, None, device, is_test=True)
        history['train_acc'].append(train_acc)
        history['test_acc'].append(test_acc)
        history['train_loss'].append(train_loss)
        history['test_loss'].append(test_loss)
        print(f"Epoch {epoch+1}: Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}")
    end = time.time()

    # Визуализация feature maps и градиентов
    plot_feature_maps(model, name)
    xb, yb = next(iter(train_loader))
    xb, yb = xb.to(device), yb.to(device)
    optimizer.zero_grad()
    out = model(xb)
    loss = criterion(out, yb)
    loss.backward()
    plot_grad_flow(model.named_parameters())

    results[name] = {
        'history': history,
        'train_time': end - start,
        'params': count_parameters(model)
    }

# Вывод результатов
print("\Итоги:")
for name, r in results.items():
    acc = r['history']['test_acc'][-1]
    print(f"{name}: точность обучения={acc:.4f}, время обучения={r['train_time']:.2f}s, параметры={r['params']}")
```
